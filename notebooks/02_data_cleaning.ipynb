# 02 - Data cleaning
**Objetivo:**  
Realizar limpeza e padronização dos dados:
- Alterar tipos de coluna
- Renomear coluna
- Tratar valoress NaN
- Salvar dataset limpo 

```python
# CÉLULA 1 - CARREGAMENTO

import pandas as pd

df = pd.read_csv("../data/raw/raw_appointments.csv")

df.head()
```

```python
# CÉLULA 2 - ALTERAR TIPO DA COLUNA 'appointment_date'

df['appointment_date'] = pd.to_datetime(df['appointment_date'], dayfirst=True)
```

```python
# CÉLULA 3 - RENOMEAR 'price'

df = df.rename(columns={'price': 'appointment_price'}) 
```

```python
# CÉLULA 4 - TRATAR VALORES NaN

df['appointment_price'] = df['appointment_price'].fillna(0)
```

```python
# CÉLULA 5 - VALIDAÇÃO

df.info()
```

```python
# CÉLULA 6 - SALVAR DATASET LIMPO

df.to_csv("../data/processed/appointments_clean.csv", index=False)
```
