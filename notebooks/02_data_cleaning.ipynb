# 02 - Data cleaning
**Objetivo:**  
Realizar limpeza e padronização dos dados:
- Alterar tipos de coluna
- Renomear coluna
- Tratar valoress NaN
- Salvar dataset limpo 

```python
# ======================
# Carregar CSV raw
# ======================

import pandas as pd

df = pd.read_csv("../data/raw/appointments_raw.csv")

df.head()
```

```python
# ===========================================
# Alterar tipo da coluna 'appointment_date'
# ===========================================

df['appointment_date'] = pd.to_datetime(df['appointment_date'], dayfirst=True)
```

```python
# =============================
# Renomear coluna 'price'
# =============================

df = df.rename(columns={'price': 'appointment_price'}) 
```

```python
# =============================
# Tratar valores nulos
# =============================

df['appointment_price'] = df['appointment_price'].fillna(0)
```

```python
# =============================
# Validação
# =============================

df.info()
```

```python
# =============================
# Salvar dataset limpo
# =============================

df.to_csv("../data/processed/appointments_clean.csv", index=False)
```
