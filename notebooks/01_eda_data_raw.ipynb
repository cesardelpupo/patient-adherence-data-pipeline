# 01 - Exploratory data analysis (EDA)

**Objetivo:**  
Realizar análise exploratória inicial no dataset bruto para verificar:
- Estrutura dos dados
- Qualidade dos dados
- Tipos de variáveis
- Possíveis problemas de limpeza

```python
# ======================
# Carregar CSV raw
# ======================

import pandas as pd

df = pd.read_csv("../data/raw/appointments_raw.csv")

df.head()
```

```python
# ======================
# Qualidade dos dados
# ======================

print("== SHAPE ==")
print(df.shape)

print("\n== COLUNAS ==")
print(df.columns.tolist())

print("\n== TIPOS ==")
print(df.dtypes)

print("\n=== NULLS ===")
print(df.isnull().sum())

print("\n=== DUPLICATAS ===")
print(f"Duplicatas: {df.duplicated().sum()}")

print("\n=== TIPOS ===")
print(df.dtypes)

print("\n=== VALORES ÚNICOS ===")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} únicos")
```
