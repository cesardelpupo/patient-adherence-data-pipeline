# 03 - Feature Engineering
**Objetivos:**  
Criar features engineering e exportar dataset analítico.

**Features:**
- attended_flag
- is_package

**Colunas temporais:**
- month
- month_date (tipo datetime)
- weekday
- weekday_num (para classificação dos dias da semana)

```python
# ======================
# Carregar CSV limpo
# ======================

import pandas as pd

df = pd.read_csv("../data/processed/appointments_clean.csv", parse_dates=['appointment_date'])

df.head()
```

```python
# ======================
# Feature engineering
# ======================

# Attendance flag
# Attended = 1 | Missed = 0
df['attended_flag'] = ( df['appointment_status'] == 'Attended').astype(int)

# Payment model flag
# MonthlyPackage = 1 | PerSession = 0
df['is_package'] = ( df['payment_type'] == 'MonthlyPackage' ).astype(int)

# ======================
# Time-based features
# ======================

# Year-month period (YYYY - MM)
df['month'] = df['appointment_date'].dt.to_period("M") 

# Month as datetime (YYYY - MM)
df['month_date'] = df['month'].dt.to_timestamp() 

# weekday name
df['weekday'] = df['appointment_date'].dt.day_name() 

# weekday number
# (1=Monday ... 7=Sunday)
df['weekday_num'] = df['appointment_date'].dt.dayofweek + 1 

df.info()
```

```python
# ================================
# Salvar CSV com novas features
# ================================

df.to_csv("../data/processed/appointments_feature.csv", index=False)
```
